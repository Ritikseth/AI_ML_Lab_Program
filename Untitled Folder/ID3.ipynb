```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../datasets/Iris.csv")
data
```

```python
x = data.drop("Species",axis=1)
y= data["Species"]
print(x.head());
print(y.head());
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =.2,random_state =3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion="entropy", max_depth=3)
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
feature_name = ["SepalLengthCm"  ,"SepalWidthCm"  ,"PetalLengthCm"  ,"PetalWidthCm"]
target_name = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(10,10))
decision_tree = plot_tree(decision_tree = dt,feature_names = feature_name,class_names=target_name,filled = True,precision=4,rounded=True)
```
